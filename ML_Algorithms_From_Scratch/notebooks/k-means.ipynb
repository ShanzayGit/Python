```python
#Reading and saving into excel file
import numpy as np
import pandas as pd
import random
import math


import sys
import os
sys.path.append(os.path.abspath(".."))
from src.kMeans import kmeans


# Read Excel file
df = pd.read_excel("../data/k_mean_clustering.xlsx", engine="openpyxl")

# Extract x and y columns
data = df[['x', 'y']].values

k = 3   
centroids, cluster_assignments = kmeans(data, k)

# Write cluster assignments back to excel sheet
df.cluster = cluster_assignments
# Save updated Excel file
df.to_excel("k_mean_clustering.xlsx", index=False)

print("Clustering complete. Results written to k_mean_clustering.xlsx")
print("CENTROIDS :\n",centroids,"\nCLUSTER ASSIGNMENT \n",cluster_assignments)
```
